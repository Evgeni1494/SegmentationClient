```python
import pandas as pd

df= pd.read_csv('df_pr_kmean')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

%matplotlib inline

columns = ['Total','gros_client', 'frequence_achat']
data = df[columns]

inertias = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
sns.lineplot(x=range(1, 11), y=inertias, marker='o')
plt.xlabel('Nombre de clusters')
plt.ylabel('Inertie')
plt.title('MÃ©thode du coude')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Ãtape 1 : Charger les donnÃ©es et sÃ©lectionner les colonnes numÃ©riques

numeric_columns = ['Total','gros_client', 'frequence_achat']
numeric_data = data[numeric_columns]

# Ãtape 2 : PrÃ©traitement des donnÃ©es
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)

# Ãtape 3 : Choix du nombre de clusters en utilisant la mÃ©thode du coude
inertia = []
k_values = range(1, 10)  # Testez diffÃ©rents nombres de clusters
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Tracer le graphique de la mÃ©thode du coude
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Nombre de clusters (k)')
plt.ylabel('Inertie')
plt.title('MÃ©thode du coude')
plt.show()
```
